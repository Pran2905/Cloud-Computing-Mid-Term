```python
# ============================================================
# KLASIFIKASI KLAKSON & SIRINE - ESP32 HELM
# OTOMATIS DETEKSI FOLDER (APAPUN NAMA ZIP NYA)
# ============================================================

# 1. INSTALL LIBRARY
!pip install tensorflow librosa soundfile audiomentations -q

# 2. IMPORT LIBRARY
import tensorflow as tf
import librosa
import numpy as np
import os
import zipfile
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras import models, layers
from google.colab import files
import audiomentations as A
import matplotlib.pyplot as plt

print(f"TensorFlow version: {tf.__version__}")

# 3. UPLOAD DATASET
print("\n=== UPLOAD DATASET ===")
print("Upload file ZIP yang berisi folder train/ dan test/")
print("Struktur folder harus:")
print("  train/")
print("    horn/")
print("    siren/")
print("    noise/")
print("  test/")
print("    horn/")
print("    siren/")
print("    noise/")
print("\nSilakan upload file ZIP Anda:")

uploaded = files.upload()

# Ekstrak file ZIP (apapun nama filenya)
for filename in uploaded.keys():
    if filename.endswith('.zip'):
        # Ekstrak ke /content/dataset_temp dulu
        with zipfile.ZipFile(filename, 'r') as zip_ref:
            zip_ref.extractall('/content/dataset_temp')
        print(f"✅ {filename} berhasil diekstrak")

# 4. CARI FOLDER TRAIN DAN TEST (AUTO-DETECT)
print("\n=== MENDETEKSI STRUKTUR FOLDER ===")

# Cari folder 'train' dan 'test' di dalam hasil ekstraksi
train_path = None
test_path = None

for root, dirs, files in os.walk('/content/dataset_temp'):
    if 'train' in dirs:
        train_path = os.path.join(root, 'train')
    if 'test' in dirs:
        test_path = os.path.join(root, 'test')

# Jika tidak ditemukan train/test, coba langsung gunakan root
if train_path is None:
    # Mungkin struktur langsung: dataset_audio/train
    if os.path.exists('/content/dataset_temp/train'):
        train_path = '/content/dataset_temp/train'
    elif os.path.exists('/content/dataset_temp/dataset_audio/train'):
        train_path = '/content/dataset_temp/dataset_audio/train'
    else:
        # Coba cari folder yang mengandung horn/siren/noise
        for item in os.listdir('/content/dataset_temp'):
            item_path = os.path.join('/content/dataset_temp', item)
            if os.path.isdir(item_path):
                subdirs = os.listdir(item_path)
                if 'horn' in subdirs or 'siren' in subdirs or 'noise' in subdirs:
                    train_path = item_path
                    print(f"📁 Folder train terdeteksi: {train_path}")
                    break

if test_path is None:
    if os.path.exists('/content/dataset_temp/test'):
        test_path = '/content/dataset_temp/test'
    elif os.path.exists('/content/dataset_temp/dataset_audio/test'):
        test_path = '/content/dataset_temp/dataset_audio/test'
    else:
        # Coba cari folder test yang berbeda dari train
        for item in os.listdir('/content/dataset_temp'):
            item_path = os.path.join('/content/dataset_temp', item)
            if os.path.isdir(item_path) and item_path != train_path:
                subdirs = os.listdir(item_path)
                if 'horn' in subdirs or 'siren' in subdirs or 'noise' in subdirs:
                    test_path = item_path
                    print(f"📁 Folder test terdeteksi: {test_path}")
                    break

# Jika masih tidak ditemukan, gunakan folder yang sama untuk train dan test
if train_path is None:
    print("❌ ERROR: Tidak dapat menemukan folder train!")
    print("Isi direktori:")
    !ls -la /content/dataset_temp
    raise Exception("Folder train tidak ditemukan")

if test_path is None:
    print("⚠️ Folder test tidak ditemukan, menggunakan data train untuk testing (tidak ideal)")
    test_path = train_path

print(f"✅ Train path: {train_path}")
print(f"✅ Test path: {test_path}")

# 5. DATA AUGMENTATION
augmenter = A.Compose([
    A.AddGaussianNoise(min_amplitude=0.001, max_amplitude=0.015, p=0.5),
    A.PitchShift(min_semitones=-4, max_semitones=4, p=0.5),
    A.TimeStretch(min_rate=0.8, max_rate=1.25, p=0.5),
])

def augment_audio(audio, sr):
    try:
        augmented = augmenter(samples=audio, sample_rate=sr)
        return augmented
    except:
        return audio

# 6. EKSTRAKSI MFCC TIME-SERIES
def extract_mfcc_ts(file_path, duration=1.0, sr=16000, n_mfcc=40, augment=False):
    try:
        audio, sr = librosa.load(file_path, sr=sr, duration=duration)

        target_len = sr * duration
        if len(audio) < target_len:
            audio = np.pad(audio, (0, target_len - len(audio)))

        if augment:
            audio = augment_audio(audio, sr)

        mfcc = librosa.feature.mfcc(
            y=audio,
            sr=sr,
            n_mfcc=n_mfcc,
            n_fft=512,
            hop_length=256
        )

        mfcc = mfcc.T

        if mfcc.shape[0] > 32:
            mfcc = mfcc[:32, :]
        elif mfcc.shape[0] < 32:
            pad_needed = int(32 - mfcc.shape[0])
            pad_width = ((0, 32 - mfcc.shape[0]), (0, 0))
            mfcc = np.pad(mfcc, pad_width, mode='constant')

        return mfcc
    except Exception as e:
        print(f"Error: {e}")
        return None

# 7. LOAD DATA
def load_data_from_folder(data_path, is_train=True):
    X, y = [], []
    for label in os.listdir(data_path):
        label_path = os.path.join(data_path, label)
        if os.path.isdir(label_path):
            print(f"  Memuat {label}...")
            for file in os.listdir(label_path):
                if file.endswith('.wav'):
                    features = extract_mfcc_ts(
                        os.path.join(label_path, file),
                        augment=is_train
                    )
                    if features is not None:
                        X.append(features)
                        y.append(label)
    return np.array(X), np.array(y)

print("\n=== MEMUAT DATA TRAIN ===")
X_train, y_train = load_data_from_folder(train_path, is_train=True)
print(f"Train data shape: {X_train.shape}")

print("\n=== MEMUAT DATA TEST ===")
X_test, y_test = load_data_from_folder(test_path, is_train=False)
print(f"Test data shape: {X_test.shape}")

if len(X_train) == 0 or len(X_test) == 0:
    print("❌ ERROR: Dataset tidak ditemukan!")
    print("\nIsi folder:")
    !ls -la {train_path}
else:
    # 8. ENCODE LABEL
    encoder = LabelEncoder()
    y_train_encoded = encoder.fit_transform(y_train)
    y_test_encoded = encoder.transform(y_test)

    y_train_cat = tf.keras.utils.to_categorical(y_train_encoded)
    y_test_cat = tf.keras.utils.to_categorical(y_test_encoded)

    print(f"\nLabel mapping: {dict(zip(encoder.classes_, range(len(encoder.classes_))))}")

    # 9. BANGUN MODEL CNN
    input_shape = (X_train.shape[1], X_train.shape[2], 1)
    num_classes = len(encoder.classes_)

    model = models.Sequential([
        layers.Input(shape=input_shape),
        layers.Conv2D(16, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.GlobalAveragePooling2D(),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(num_classes, activation='softmax')
    ])

    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
    model.summary()

    # 10. RESHAPE
    X_train_cnn = X_train.reshape(X_train.shape[0], X_train.shape[1], X_train.shape[2], 1)
    X_test_cnn = X_test.reshape(X_test.shape[0], X_test.shape[1], X_test.shape[2], 1)

    # 11. TRAINING
    print("\n=== MEMULAI TRAINING ===")
    history = model.fit(
        X_train_cnn, y_train_cat,
        validation_split=0.2,
        epochs=100,
        batch_size=16,
        verbose=1
    )

    # 12. PLOT
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
    ax1.plot(history.history['accuracy'], label='Train')
    ax1.plot(history.history['val_accuracy'], label='Val')
    ax1.set_title('Accuracy')
    ax1.legend()
    ax2.plot(history.history['loss'], label='Train')
    ax2.plot(history.history['val_loss'], label='Val')
    ax2.set_title('Loss')
    ax2.legend()
    plt.tight_layout()
    plt.show()

    # 13. EVALUASI
    loss, accuracy = model.evaluate(X_test_cnn, y_test_cat)
    print(f"\n✅ Test Accuracy: {accuracy:.4f}")

    # 14. KONVERSI KE TFLite
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    tflite_model = converter.convert()

    with open('model.tflite', 'wb') as f:
        f.write(tflite_model)

    print(f"✅ Model size: {len(tflite_model)} bytes")
    files.download('model.tflite')
    print("\n✅ Selesai! model.tflite sudah di-download")
```

```python
# ============================================================
# KONVERSI FULL INT8 UNTUK ESP32-C3
# ============================================================

def representative_dataset():
    # Gunakan beberapa ratus sampel training sebagai kalibrasi
    max_samples = min(300, len(X_train_cnn))

    for i in range(max_samples):
        sample = X_train_cnn[i:i+1].astype(np.float32)
        yield [sample]

converter = tf.lite.TFLiteConverter.from_keras_model(model)

converter.optimizations = [tf.lite.Optimize.DEFAULT]
converter.representative_dataset = representative_dataset

converter.target_spec.supported_ops = [
    tf.lite.OpsSet.TFLITE_BUILTINS_INT8
]

converter.inference_input_type = tf.int8
converter.inference_output_type = tf.int8

tflite_model = converter.convert()

with open("model_int8.tflite", "wb") as f:
    f.write(tflite_model)

print("✅ INT8 model size:", len(tflite_model), "bytes")
```

```python
# ============================================================
# BUAT model_data.h
# ============================================================

from google.colab import files # Re-import files module

model_bytes = open("model_int8.tflite", "rb").read()

with open("model_data.h", "w") as f:
    f.write("#pragma once\n")
    f.write("#include <stdint.h>\n\n")
    f.write("alignas(16) const unsigned char model_tflite[] = {\n")

    for i, value in enumerate(model_bytes):
        if i % 12 == 0:
            f.write("  ")

        f.write(f"0x{value:02x}")

        if i < len(model_bytes) - 1:
            f.write(", ")

        if i % 12 == 11:
            f.write("\n")

    if len(model_bytes) % 12 != 0:
        f.write("\n")

    f.write("};\n\n")
    f.write(
        f"const unsigned int model_tflite_len = "
        f"{len(model_bytes)};\n"
    )

files.download("model_data.h")
files.download("model_int8.tflite")
```
